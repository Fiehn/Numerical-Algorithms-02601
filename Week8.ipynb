```python
dxdt = lambda t, x, y: -x + y + np.sin(t)
dydt = lambda t, x, y: -x - y
```

```python
x0 = 0
y0 = 0
```

```python
# Convert to system
f = f2(dxdt,dydt)
```

```python
f(0,[1,2])
```

```python
t, X = RungeKutta4System(f, [0,10], [x0, y0], 20)
```

```python
x_true = lambda t: 0.2*np.exp(-t)*np.cos(t) - 0.4*np.exp(-t)*np.sin(t) + 0.6*np.sin(t) - 0.2*np.cos(t)
y_true = lambda t: -0.2*np.exp(-t)*np.sin(t) - 0.4*np.exp(-t)*np.cos(t) - 0.2*np.sin(t) + 0.4*np.cos(t)
t_true = np.linspace(0,10,100)
```

```python
plt.plot(t, X[:,0], label='x')
plt.plot(t, X[:,1], label='y')
plt.plot(t_true, x_true(t_true), label='x_true')
plt.plot(t_true, y_true(t_true), label='y_true');
```

```python
etx = np.abs(x_true(t) - X[:,0])
ety = np.abs(y_true(t) - X[:,1])
plt.plot(t, etx, label='x')
plt.plot(t, ety, label='y');
```

```python
def test(tspan, n):
    t, X = RungeKutta4System(f, tspan, [x0, y0], n)
    etx = np.abs(x_true(t) - X[:,0])
    ety = np.abs(y_true(t) - X[:,1])
    plt.plot(t, etx, label='x')
    plt.plot(t, ety, label='y');
```

```python
test([0,100], 50)
```
