```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def TaylorOrder2(f, df, x0, a, b, n):
    """ Calculates the Taylor polynomial of order 2 for f on [a,b] using n steps
        f: function as lambda function
        df: derivative of f as lambda function
        x0: point of expansion
        a: lower bound
        b: upper bound
        n: number of steps
    """
    h = (b-a)/n
    t = np.linspace(a,b,n+1) 
    x = np.zeros(n+1)
    x[0] = x0
    for i in range(n):
        dx = f(t[i], x[i])
        ddx = df(t[i], x[i], dx)
        x[i+1] = x[i] + h * (dx + h/2 * (ddx))
    return t, x
    
```

```python
f = lambda t, x: t + x
df = lambda t, x, dx: 1 + dx
TaylorOrder2(f, df, 1.0, 0.0, 1.0, 4)
```

```python
fejl = []
for i in [16,32,64,128,256,512]:
    a = TaylorOrder2(f, df, 1.0, 0.0, 1.0, i)
    fejl.append(a[1][np.where(a[0] == 0.5)])
```
